```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from skimage.io import imread
from skimage.color import rgb2gray

image = imread('https://img.freepik.com/free-vector/cute-happy-smiling-child-isolated-white_1308-32243.jpg?w=360&t=st=1729530333~exp=1729530933~hmac=5100e3119af4ed43bba0cda33a62fffcb1ba49a1e5bf6edf4717c0c75085ba7a')
gray_image = rgb2gray(image)

X = gray_image.reshape(gray_image.shape[0], -1)

pca_5 = PCA(n_components=5)
X_pca_5 = pca_5.fit_transform(X)
X_reconstructed_5 = pca_5.inverse_transform(X_pca_5)

pca_20 = PCA(n_components=20)
X_pca_20 = pca_20.fit_transform(X)
X_reconstructed_20 = pca_20.inverse_transform(X_pca_20)

pca_50 = PCA(n_components=50)
X_pca_50 = pca_50.fit_transform(X)
X_reconstructed_50 = pca_50.inverse_transform(X_pca_50)

pca_100 = PCA(n_components=100)
X_pca_100 = pca_100.fit_transform(X)
X_reconstructed_100 = pca_100.inverse_transform(X_pca_100)

fig, axes = plt.subplots(1, 5, figsize=(15, 5))

axes[0].imshow(gray_image, cmap='gray')
axes[0].set_title('Original Image')

axes[1].imshow(X_reconstructed_5.reshape(gray_image.shape), cmap='gray')
axes[1].set_title('5 Components')

axes[2].imshow(X_reconstructed_20.reshape(gray_image.shape), cmap='gray')
axes[2].set_title('20 Components')

axes[3].imshow(X_reconstructed_50.reshape(gray_image.shape), cmap='gray')
axes[3].set_title('50 Components')

axes[4].imshow(X_reconstructed_100.reshape(gray_image.shape), cmap='gray')
axes[4].set_title('100 Components')

for ax in axes:
    ax.axis('off')

plt.tight_layout()
plt.show()


```
